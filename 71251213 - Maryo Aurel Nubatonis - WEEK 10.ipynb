```python
file1 = input("Nama file 1: ")
file2 = input("Nama file 2: ")

try:
    f1 = open(file1)
    f2 = open(file2)

    for i, (baris1, baris2) in enumerate(zip(f1, f2), start=1):
        if baris1.strip() != baris2.strip():
            print(f"Baris {i}:")
            print("File 1:", baris1.strip())
            print("File 2:", baris2.strip())
            print("-" * 25)

    f1.close()
    f2.close()

except:
    print("File tidak ditemukan!")
```

```python
filename = input("nama file1: ")

try:
    file = open(filename)

    for line in file:
        data = line.strip().split("||")
        
        soal = data[0].strip()
        jawaban_benar = data[1].strip()

        print(soal)
        jawab = input("Jawab: ")

        if jawab.lower() == jawaban_benar.lower():
            print("Jawaban benar!\n")
        else:
            print("Jawaban salah!\n")

    file.close()

except:
    print("File tidak ditemukan!")
```
